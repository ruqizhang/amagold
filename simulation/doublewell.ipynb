```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import time
import pdb
import matplotlib
# import some stuff
from __future__ import division, print_function

from scipy.stats import gaussian_kde
from scipy.stats import norm, uniform, cauchy
import scipy.stats as st
import os
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
sns.set_style("whitegrid")
import scipy.io as sio
```

```python
# Double-well
res = .1
y = np.arange(-6,6, res)
z= np.arange(-6,6, 0.05)
sigma2 = 0.3
sigma = sigma2**.5
sigma_prime = 0.5**.5
U = lambda x: (x + 4)*(x + 1)*(x - 1)*(x - 3)/14 + 0.5
p = np.exp( - U(y) );
p = p / sum(p) / res;

mat_contents = sio.loadmat('sghmc_doublewell.mat') 
sghmc =  mat_contents['samples']
sghmc = sghmc.reshape(1,-1)[0]
plt.figure(1)
sns.kdeplot(sghmc, shade = True, label = 'SGHMC', color = 'b')
plt.plot(y, p, label="Truth", color = 'k',lw=2)
plt.legend(fontsize=17)
plt.savefig('doublewell_sghmc.pdf')

mat_contents = sio.loadmat('amagold_doublewell.mat') 
ima_sghmc =  mat_contents['samples']
ima_sghmc = ima_sghmc.reshape(1,-1)[0]
plt.figure(2)
sns.kdeplot(ima_sghmc, shade = True, label = 'AMAGOLD', color = 'b')
plt.plot(y, p, label="Truth", color = 'k',lw=2)
plt.legend(fontsize=17)
plt.savefig('doublewell_amagold.pdf')
```
